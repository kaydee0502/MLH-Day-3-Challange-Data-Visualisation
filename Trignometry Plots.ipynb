# Trignometry Plots

> Initialize Degrees : -360 to 360

```python
import numpy as np
import matplotlib.pyplot as plt

degree = np.linspace(-2*np.pi,2*np.pi,1000)
labels = ["-2\u03C0","-\u03C0","0","\u03C0","2\u03C0"]
labels_deg = [-2*np.pi,-np.pi,0,np.pi,2*np.pi]
```

### Sin

```python
plt.plot(degree,np.sin(degree))
_ = plt.xticks(labels_deg,labels)
```

### Cos

```python
plt.plot(degree,np.cos(degree))
_ = plt.xticks(labels_deg,labels)
```

### Tan

```python
plt.scatter(degree,np.tan(degree),s=2.5)
plt.ylim(-5,5)
_ = plt.xticks(labels_deg,labels)
```

### Cot

```python
plt.scatter(degree,1/np.tan(degree),c="red",s=2.5)
plt.ylim(-5,5)
_ = plt.xticks(labels_deg,labels)
```

## Tan + Cot

```python
plt.scatter(degree,np.tan(degree),s=2.5)
plt.scatter(degree,1/np.tan(degree),c="red",s=2.5)
plt.ylim(-5,5)
_ = plt.xticks(labels_deg,labels)
```

### Sec

```python
plt.scatter(degree,1/np.cos(degree),c="red",s=2.5)
plt.ylim(-5,5)
_ = plt.xticks(labels_deg,labels)
```

## Cos + Sec

```python
plt.plot(degree,np.cos(degree))
plt.scatter(degree,1/np.cos(degree),c="red",s=2.5)
plt.ylim(-5,5)
_ = plt.xticks(labels_deg,labels)
```

### Cosec

```python
plt.scatter(degree,1/np.sin(degree),s=2.5,c="red")
plt.ylim(-5,5)
_ = plt.xticks(labels_deg,labels)
```

## Cosec + Sin

```python
plt.plot(degree,np.sin(degree))
plt.scatter(degree,1/np.sin(degree),s=2.5,c="red")
plt.ylim(-5,5)
_ = plt.xticks(labels_deg,labels)
```
