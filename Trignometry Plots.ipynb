# Trignometry Plots

> Initialize Degrees : -360 to 360

```python
import numpy as np
import matplotlib.pyplot as plt

degree = np.linspace(-2*np.pi,2*np.pi,100)
```

### Sin

```python
plt.plot(degree,np.sin(degree))
```

### Cos

```python
plt.plot(degree,np.cos(degree))
```

### Tan

```python
plt.scatter(degree,np.tan(degree),s=2.5)
plt.ylim(-5,5)
```

### Cot

```python
plt.scatter(degree,1/np.tan(degree),c="red",s=2.5)
plt.ylim(5,-5)
```

## Tan + Cot

```python
plt.scatter(degree,np.tan(degree),s=2.5)
plt.scatter(degree,1/np.tan(degree),c="red",s=2.5)
plt.ylim(5,-5)
```

### Sec

```python
plt.scatter(degree,1/np.cos(degree),c="red",s=2.5)
plt.ylim(5,-5)
```

## Cos + Sec

```python
plt.plot(degree,np.cos(degree))
plt.scatter(degree,1/np.cos(degree),c="red",s=2.5)
plt.ylim(5,-5)
```

### Cosec

```python
plt.scatter(degree,1/np.sin(degree),s=2.5,c="red")
plt.ylim(5,-5)
```

## Cosec + Sin

```python
plt.plot(degree,np.sin(degree))
plt.scatter(degree,1/np.sin(degree),s=2.5,c="red")
plt.ylim(5,-5)
```
